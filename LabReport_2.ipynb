```python
import pandas as pd
import numpy as np
from sklearn.datasets import load_iris
from sklearn.model_selection import train_test_split
from sklearn.preprocessing import MinMaxScaler
from sklearn.neighbors import KNeighborsClassifier
from sklearn.metrics import accuracy_score, precision_score, recall_score, f1_score
```

```python
iris = load_iris()
X = pd.DataFrame(iris.data, columns=iris.feature_names)
y = iris.target
```

```python
X.fillna(X.mean(), inplace=True)
```

```python
scaler = MinMaxScaler()
X_scaled = scaler.fit_transform(X)
```

```python
split_ratios = [0.50, 0.40, 0.30, 0.20] # Corresponding to 50:50, 60:40, 70:30, 80:20
k_values = [1, 3, 5, 7, 9]

results = []

for test_size in split_ratios:
    X_train, X_test, y_train, y_test = train_test_split(X_scaled, y, test_size=test_size, random_state=42)
```

```python
for k in k_values:
    knn = KNeighborsClassifier(n_neighbors=k)
    knn.fit(X_train, y_train)
    y_pred = knn.predict(X_test)
    acc = accuracy_score(y_test, y_pred)
    prec = precision_score(y_test, y_pred, average='weighted')
    rec = recall_score(y_test, y_pred, average='weighted')
    f1 = f1_score(y_test, y_pred, average='weighted')

    results.append({
        'Split Ratio': f"{int((1-test_size)*100)}:{int(test_size*100)}",
        'k_value': k,
        'Accuracy': acc,
        'Precision': prec,
        'Recall': rec,
        'F1-Score': f1
    })
```

```python
results_df = pd.DataFrame(results)
print(results_df.to_string(index=False))
```

```python
best_row = results_df.loc[results_df['Accuracy'].idxmax()]
print(f"\nOptimal Configuration: {best_row['Split Ratio']} split with k={best_row['k_value']}")
```
